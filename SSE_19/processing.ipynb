```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1 = pd.read_csv('t1.csv')
b1 = t1['beta']
I = t1['I_k']
t2 = pd.read_csv('t2.csv')
b2 = t2['beta']
V = t2['V_k']
```

```python
t1
```

```python
t2
```

```python
plt.figure(figsize=(10,6))

plt.plot(I, b1)
plt.scatter(I, b1)

plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=1)
plt.grid()

plt.xlabel(r'$I_k$, mA')
plt.ylabel(r'$\beta_N$')
plt.savefig('pictures/p1.png', dpi=400)
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.plot(V, b2)
plt.scatter(V, b2)

plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth=1)
plt.grid()

plt.xlabel(r'$V_k$, V')
plt.ylabel(r'$\beta_N$')
plt.savefig('pictures/p2.png', dpi=400)
plt.show()
```
